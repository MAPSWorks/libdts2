```python
import math
import numpy

maxlat = 70
nc = "0 0 0 0"
ec = "0 0 0 255"

def n2s(n):
    return str(n.item(0)) + " " + str(n.item(1)) + " " + nc

def e2s(e):
    return str(e[0]) + " " + str(e[1]) + " " + ec

def cartesianFromSpherical(sp):
    sinTheta = math.sin(sp.item(0));
    x = sinTheta * math.cos(sp.item(1));
    y = sinTheta * math.sin(sp.item(1));
    z = math.cos(sp.item(0));
    return numpy.matrix([x,y,z])

#from cartesian
def sphericalFromCartesian(c):
    theta = math.acos(c.item(2));
    phi = math.atan(c.item(1) / c.item(0));
    if c.item(0) < 0:
        if c.item(1) < 0:
            phi -= math.pi
        else:
            phi += math.pi
    return numpy.matrix([theta, phi])

def sphericalFromGeo(g):
    lat = g.item(0)
    lon = g.item(1)
    theta = math.pi/180.0 * (90.0 - lat);
    phi = 0.0
    if lon >= 0:
        phi = ((lon * math.pi)/ 180)
    else:
       phi =((lon + 360) * math.pi)/ 180;
    return numpy.matrix([theta, phi])

#from geo
def cartesianFromGeo(g):
    return cartesianFromSpherical( sphericalFromGeo(g) );
    
#from cartesian
def geoFromCartesian(c):
    sp = sphericalFromCartesian(c)
    theta = sp.item(0)
    phi = sp.item(1)
    lat = 90 - theta / math.pi *  180;
    lon = phi / math.pi * 180;
    if phi > math.pi:
        lon = lon - 90;
    return numpy.matrix([lat, lon])

def arc_trace(gs, gt, steps, nodes, edges):
    cs = cartesianFromGeo(gs)
    ct = cartesianFromGeo(gt)
    dc = ct-cs
    nodes.append(gs)
    for i in range(1, steps+1):
        t = i/float(steps)
        cn = cs + t*dc
        cn /= numpy.linalg.norm(cn)
        nodes.append(geoFromCartesian(cn))
        edges.append([len(nodes)-2, len(nodes)-1])
        
steps = 1000
nodes = []
edges = []

t1 = numpy.matrix([maxlat, 0])
t2 = numpy.matrix([maxlat, 120])
t3 = numpy.matrix([maxlat, -120])
sp = numpy.matrix([-90, 0])

arc_trace(sp, t1, steps, nodes, edges)
arc_trace(sp, t2, steps, nodes, edges)
arc_trace(sp, t3, steps, nodes, edges)
arc_trace(t1, t2, steps, nodes, edges)
arc_trace(t2, t3, steps, nodes, edges)
arc_trace(t3, t1, steps, nodes, edges)

fo = open("dts_initial_triang.sg", 'w')

fo.write(str(len(nodes)) + "\n")
fo.write(str(len(edges)) + "\n")
fo.write("0\n")
for n in nodes:
    fo.write(n2s(n) + "\n")
for e in edges:
    fo.write(e2s(e) + "\n")
                
fo.flush()
fo.close()
```
